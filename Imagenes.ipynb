```python
def find_pixels_with_color(image, colors_RGB, epsilon=0):
    # Retorna un arreglo de 2 dimensiones con las coordenadas de los pixeles tienen los colores pedidos con un error epsilon
    # donde la primera lista la fila y la segunda la columna
    # image: nombre del archivo de la imagen
    # colors_RGB un arreglo de longitud 3 listando los colores
    img = cv2.imread(image)
    r=2
    g=1
    b=0
    r_query = colors_RGB[0]
    g_query = colors_RGB[1]
    b_query = colors_RGB[2]
    pixels=np.where((img[:,:,r] >= r_query-epsilon)& (img[:,:,r]<= r_query+epsilon) & (img[:,:,g] >= g_query-epsilon)& (img[:,:,g] <= g_query+epsilon) & (img[:,:,b]>= b_query-epsilon) & (img[:,:,b]<= b_query+epsilon))
    return(pixels)
```

```python
def draw_circle_around_pixels(imagen, pixels, radius=5):
    # imagen = nombre del archivo a cargar
    # coordinates = coordenadas del centro del circulo (fila, columna)
    # radius =radio del circulo
    # Dibuja un circulo amarillo de radio=radius alrededor del centro dado por coordinates
    img = plt.imread(imagen)

    x=pixels[1] # primero ponemos la segunda coordenada del pixel
    y=pixels[0] # luego la primera coordenada
    
    # Create a figure. Equal aspect so circles look circular
    fig,ax = plt.subplots(1)
    ax.set_aspect('equal')
    
    # Show the image
    ax.imshow(img)
    for xx,yy in zip(x,y):
        circ = Circle((xx,yy),radius=radius, fill=False, linewidth=2, color="yellow")
        ax.add_patch(circ)
```

```python
def identify_pixels_with_color(imagen, colors_RGB, epsilon=0, radius=5):
    pixels=find_pixels_with_color(imagen, colors_RGB, epsilon)
    if (len(pixels[0])==0):
        print("No existen pixels con ese color")
        return
    draw_circle_around_pixels(imagen, pixels, radius=radius)
#     for k in range(len(pixels[0])):
#         x=pixels[0][k]
#         y=pixels[1][k]
#         draw_circle_around_pixel(imagen, [x, y], radius)    
```
